```python
import os
from PIL import Image
import shutil
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Dropout, Conv2DTranspose, concatenate, BatchNormalization, Activation, UpSampling2D
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
# Directory paths
base_dirs = ["./Training", "./Testing"]
mask_dirs = ["./New_Mask_Training", "./New_Mask_Testing"]
categories = ['glioma_tumor', 'meningioma_tumor', 'no_tumor', 'pituitary_tumor']

# Labels dictionary
labels_dict = {'glioma_tumor': 0, 'meningioma_tumor': 0, 'no_tumor': 1, 'pituitary_tumor': 0}

# Storage lists
images = []
masks = []
image_labels = []

# Loop through directories and categories
for base_dir, mask_dir in zip(base_dirs, mask_dirs):
    for category in categories:
        img_category_path = os.path.join(base_dir, category)
        mask_category_path = os.path.join(mask_dir, category)
        
        # Ensure mask directory exists
        os.makedirs(mask_category_path, exist_ok=True)
        
        mask_files = set(os.listdir(mask_category_path))
        
        for img_name in os.listdir(img_category_path):
            img_base_name, ext = os.path.splitext(img_name)
            if ext.lower() in ['.png', '.jpg', '.jpeg']:
                img_path = os.path.join(img_category_path, img_name)
                mask_path = os.path.join(mask_category_path, img_base_name + '.png')  # Constructing the mask path
                
                # Load and preprocess the image
                img = Image.open(img_path).resize((128, 128))
                img = np.array(img)
                
                # Initialize mask array as zeros
                mask = np.zeros((128, 128), dtype=np.uint8)
                
                # Load and preprocess the mask if it exists
                if img_name in mask_files:
                    mask_img = Image.open(mask_path).resize((128, 128))
                    mask = np.array(mask_img)
                    if len(mask.shape) == 3:  # Convert RGB to grayscale if necessary
                        mask = mask[:, :, 0]
                
                # Append data to lists
                images.append(img)
                masks.append(mask)
                image_labels.append(labels_dict[category])

# convert lists to numpy arrays
images = np.array(images)
masks = np.array(masks)
image_labels = np.array(image_labels)

print(f"Loaded {len(images)} images, {len(masks)} masks, and {len(image_labels)} labels.")
```

```python
# Choose a random image
idx = 450

# Prepare the image and mask
image_to_show = images[idx] if images[idx].shape[-1] == 3 else np.stack((images[idx],)*3, axis=-1)
mask_to_show = np.ma.masked_where(masks[idx] == 0, masks[idx])

# Plotting
fig, ax = plt.subplots(figsize=(6, 6))
ax.imshow(image_to_show, cmap='gray')
ax.imshow(mask_to_show, cmap='Reds', alpha=0.5)  # overlay mask in red

# Title with label
label_titles = {0: 'Tumor', 1: 'No Tumor'}  # based on labelling
ax.set_title(label_titles[image_labels[idx]])

# Legend for the mask
red_patch = mpatches.Patch(color='red', label='Mask')
plt.legend(handles=[red_patch])

plt.axis('off')
plt.show()
```
